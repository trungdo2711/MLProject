```python
file_path = r'G:\url-analysis\src\data\processed_url_update.csv'
url_data = pd.read_csv(file_path, encoding_errors='ignore')

mean_benign_part = url_data[url_data['type'] == 'benign']['number_of_part'].mean()
mean_benign_scheme = url_data[url_data['type'] == 'benign']['has_scheme'].mean()
mean_benign_levenshtein = url_data[url_data['type'] == 'benign']['normalized_levenshtein_domain'].mean()
print(f"mean value of number of part in benign: {mean_benign_part}")
print(f"mean value of number of url having scheme in benign: {mean_benign_scheme}")
print(f"mean value of levenshtein of benign url: {mean_benign_levenshtein}")
print()

mean_defacement_part = url_data[url_data['type'] == 'defacement']['number_of_part'].mean()
mean_defacement_scheme = url_data[url_data['type'] == 'defacement']['has_scheme'].mean()
mean_defacement_levenshtein = url_data[url_data['type'] == 'defacement']['normalized_levenshtein_domain'].mean()
print(f"mean value of number of part in defacement: {mean_defacement_part}")
print(f"mean value of number of url having scheme in defacement: {mean_defacement_scheme}")
print(f"mean value of levenshtein of defacement url: {mean_defacement_levenshtein}")
print()

mean_malware_part = url_data[url_data['type'] == 'malware']['number_of_part'].mean()
mean_malware_scheme = url_data[url_data['type'] == 'malware']['has_scheme'].mean()
mean_malware_levenshtein = url_data[url_data['type'] == 'malware']['normalized_levenshtein_domain'].mean()
print(f"mean value of number of part in malware: {mean_malware_part}")
print(f"mean value of number of url having scheme in malware: {mean_malware_scheme}")
print(f"mean value of levenshtein of malware url: {mean_malware_levenshtein}")
print()

mean_phishing_part = url_data[url_data['type'] == 'phishing']['number_of_part'].mean()
mean_phishing_scheme = url_data[url_data['type'] == 'phishing']['has_scheme'].mean()
mean_phishing_levenshtein = url_data[url_data['type'] == 'phishing']['normalized_levenshtein_domain'].mean()
print(f"mean value of number of part in phishing: {mean_phishing_part}")
print(f"mean value of number of url having scheme in phishing: {mean_phishing_scheme}")
print(f"mean value of levenshtein of phishing url: {mean_phishing_levenshtein}")
print()
```
